```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(123456)
```

```python
n = 1000
ts = pd.Series(np.random.randn(n), index=pd.date_range('1/1/2019', periods=n))
ts.head()
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
ts.plot(x_compat=True)
```

```python
df = pd.DataFrame(np.random.randn(n, 4),index=ts.index, columns=list('ABCD'))
```

```python
df = df.cumsum()
```

```python
plt.figure()
```

```python
df.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
```

```python
df3['A'] = pd.Series(list(range(len(df))))
```

```python
df3.plot(x='A', y='B')
```
